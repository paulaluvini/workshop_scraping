## Level 4

```python
script_list = []
if __name__ == "__main__":
    options = Options()
    driver = webdriver.Chrome(options=options)
    driver.get("https://hertie-scraping-website.vercel.app/level4")
    page= driver.find_elements(By.TAG_NAME, 'script')
    for i in page:
        print(i.get_attribute('src'))
        script_list.append(i.get_attribute('src'))
    
    for s in range(0, len(script_list)):
        if "549920bc0939d905.js" in script_list[s]:
            driver.get(script_list[s])
            text = driver.find_element(By.XPATH, "//body/pre").text
            print(text.find("flag-61"))
            print(text[489:496])
    driver.quit()
```
